# After dowloading the images per each Senegal department, open one by one and plot the deforestation status across the years (from 2001 to 2019)

#barplots of deforestation

```
img_path = "/content/drive/MyDrive/Senegal_forest forest_change/"

for filename in os.listdir(img_path):
  
  img = rasterio.open(img_path+ filename).read(2)#band 2 is the lossyear, band 1 is the treecover

  img_2019 = np.array(img, copy=True)
  
  unique, counts = np.unique(img_2019, return_counts=True)
  mydict = (dict(zip(unique, counts)))
  
  df = pd.DataFrame({'Area (square meter)':dict.values(mydict), 'years':dict.keys(mydict)})
  df = df[1:]
  df.loc[:,'Area (square meter)'] *= 22201

  
  df.plot.bar(x='years',y='Area (square meter)',rot=0,title = "Department: {}".format(os.path.splitext(filename)[0]),figsize=(10,8))
```

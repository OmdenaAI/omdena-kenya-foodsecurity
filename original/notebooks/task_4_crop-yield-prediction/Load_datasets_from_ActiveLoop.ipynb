```
def plot_bands_hist(hist):
  plt.figure(1, figsize=(10,10))
  plt.suptitle("First histogram in dataset")
  plt.subplot(331)
  plt.title("first band")
  plt.imshow(hist[0]) 
  plt.subplot(332)
  plt.title("second band")
  plt.imshow(hist[1])
  plt.subplot(333)
  plt.title("third band")
  plt.imshow(hist[2])
  plt.subplot(334)
  plt.title("Fourth band")
  plt.imshow(hist[3])
  plt.subplot(335)
  plt.title("Fifth band")
  plt.imshow(hist[4])
  plt.subplot(336)
  plt.title("Sixth band")
  plt.imshow(hist[5])
  plt.subplot(337)
  plt.title("7th band")
  plt.imshow(hist[6])
  plt.subplot(338)
  plt.title("8th band")
  plt.imshow(hist[7])
  plt.subplot(339)
  plt.title("9th band")
  plt.imshow(hist[8]) 
```

#### Load data from ActiveLoop

```
tag = "margauxmforsythe/Senegal_regions_2016_2019_week_0_46"

# Load the data
ds = Dataset(tag)
number_histograms = ds['histograms'].shape[0]
print("There are {} histograms in this dataset".format(number_histograms))
```

Load first item data

```
first_hist = ds['histograms'][0].compute() 
first_hist.shape # (9, 32, 46) : 9 bands, histgrams 32 bins x 46 timestamps (one year)
```

```
ds['yields'][0].compute() # yield value for this item (region, year, crop type)
```

```
print(ds['ndvi'][0].compute().shape) # (46,) : 46 timestamps
ds['ndvi'][0].compute()
```

```
plot_bands_hist(first_hist)
```

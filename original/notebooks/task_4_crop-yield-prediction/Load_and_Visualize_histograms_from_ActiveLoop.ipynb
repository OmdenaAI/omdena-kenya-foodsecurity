```
from hub import Dataset, schema
from matplotlib import pyplot as plt
```

#### Functions

```
def plot_histograms(hist_list, names, num_images_to_plot=5):
  i = 0
  for hist in hist_list:
    if i < num_images_to_plot:
      plt.figure(i, figsize=(10,5))
      plt.suptitle(names[i])
      plt.subplot(121)
      plt.title("first band image (temperature day)")
      plt.imshow(hist[0]) 
      plt.subplot(122)
      plt.title("second band image (temperature night)")
      plt.imshow(hist[1]) 
      i = i + 1
  plt.show()
```

#### Load data from ActiveLoop

```
tag_temperature_regions = "margauxmforsythe/Senegal_MYD11A2_Temperature_histograms_regions"
tag_reflectance_regions = "margauxmforsythe/Senegal_MOD09A1_TerraSurfaceReflectance_regions"
tag_temperature_departments = "margauxmforsythe/Senegal_MYD11A2_Temperature_histograms_departments"
tag_reflectance_departments = "margauxmforsythe/Senegal_MOD09A1_TerraSurfaceReflectance_departments"

# Load the data
ds_temperature_regions = Dataset(tag_temperature_regions)
ds_reflectance_regions = Dataset(tag_reflectance_regions)

ds_temperature_departments = Dataset(tag_temperature_departments)
ds_reflectance_departments = Dataset(tag_reflectance_departments)

################### REGIONS ########################
### Temperature ###
dataset_MYD11A2_Temperature_Senegal_regions_histograms_one_year = ds_temperature_regions["histograms_one_year"].compute()
print(dataset_MYD11A2_Temperature_Senegal_regions_histograms_one_year.shape)

dataset_MYD11A2_Temperature_Senegal_regions_histograms_june_august = ds_temperature_regions["histograms_june_august"].compute()
print(dataset_MYD11A2_Temperature_Senegal_regions_histograms_june_august.shape)

dataset_MYD11A2_Temperature_Senegal_regions_histograms_half_year = ds_temperature_regions["histograms_half_year"].compute()
print(dataset_MYD11A2_Temperature_Senegal_regions_histograms_half_year.shape)

### Reflectance ###
dataset_Senegal_MOD09A1_TerraSurfaceReflectance_regions_one_year = ds_reflectance_regions["histograms_one_year"].compute()
print(dataset_Senegal_MOD09A1_TerraSurfaceReflectance_regions_one_year.shape)

dataset_Senegal_MOD09A1_TerraSurfaceReflectance_regions_june_august = ds_reflectance_regions["histograms_june_august"].compute()
print(dataset_Senegal_MOD09A1_TerraSurfaceReflectance_regions_june_august.shape)

dataset_Senegal_MOD09A1_TerraSurfaceReflectance_regions_half_year = ds_reflectance_regions["histograms_half_year"].compute()
print(dataset_Senegal_MOD09A1_TerraSurfaceReflectance_regions_half_year.shape)


################### Departments ########################
### Temperature ###
dataset_MYD11A2_Temperature_Senegal_departments_histograms_one_year = ds_temperature_departments["histograms_one_year"].compute()
print(dataset_MYD11A2_Temperature_Senegal_departments_histograms_one_year.shape)

dataset_MYD11A2_Temperature_Senegal_departments_histograms_june_august = ds_temperature_departments["histograms_june_august"].compute()
print(dataset_MYD11A2_Temperature_Senegal_departments_histograms_june_august.shape)

dataset_MYD11A2_Temperature_Senegal_departments_histograms_half_year = ds_temperature_departments["histograms_half_year"].compute()
print(dataset_MYD11A2_Temperature_Senegal_departments_histograms_half_year.shape)

### Reflectance ###
dataset_Senegal_MOD09A1_TerraSurfaceReflectance_departments_one_year = ds_reflectance_departments["histograms_one_year"].compute()
print(dataset_Senegal_MOD09A1_TerraSurfaceReflectance_departments_one_year.shape)

dataset_Senegal_MOD09A1_TerraSurfaceReflectance_departments_june_august = ds_reflectance_departments["histograms_june_august"].compute()
print(dataset_Senegal_MOD09A1_TerraSurfaceReflectance_departments_june_august.shape)

dataset_Senegal_MOD09A1_TerraSurfaceReflectance_departments_half_year = ds_reflectance_departments["histograms_half_year"].compute()
print(dataset_Senegal_MOD09A1_TerraSurfaceReflectance_departments_half_year.shape)
```

```
names_regions = []
for i in range(len(ds_temperature_regions["names"].shape)):
  names_regions.append(ds_temperature_regions["names"][i].numpy())

names_departments = []
for i in range(len(ds_reflectance_departments["names"].shape)):
  names_departments.append(ds_reflectance_departments["names"][i].numpy())
```

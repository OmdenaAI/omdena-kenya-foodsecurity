```python
import pandas as pd
import numpy as np
#pd.set_option('display.max_rows', None)
```

### Load IPAR Maize data and Edit

```python
df = pd.read_csv('IPAR_yields_with_NDVI-Maize_2014_3.csv')
```

```python
df.dropna(inplace=True) # drop Nan rows
```

```python
df.info()
```

```python
type(df['ndvi'][0]) # ndvi data type?
```

```python
# Remove characters from string and convert to list
df['ndvi'] = df['ndvi'].str.strip(',').str.strip('[]').str.split(',')
```

```python
type(df['ndvi'][0])
```

```python
df.head(3)
```

#### Decompose ndvi list into individual values, and place in individual rows

```python
df.columns
```

```python
res = df.set_index(['KEY', 'ID_menage', 'crops_cult5', 'season_cult5', 'Area_cult5,(ha)',
       'QuantityProduced_cult5,(kg)', 'Yield', 'Department', 'Counties',
       'Village', 'gpslat', 'gpslon'])['ndvi'].apply(pd.Series).stack()
res = res.reset_index()
res.columns = ['KEY', 'ID_menage', 'crops_cult5', 'season_cult5', 'Area_cult5,(ha)',
       'QuantityProduced_cult5,(kg)', 'Yield', 'Department', 'Counties',
       'Village', 'gpslat', 'gpslon', 'ndvi_num','ndvis']
res
```

```python
# remove rows with 'problem' formats

for i in range(1, len(res['ndvi_num'])):
    get_index = res[ res['ndvi_num'] == i*5 ].index
    res.drop(get_index, inplace = True)
res.head(16)
```

#### output file for merge with weather data

```python
res.to_csv('ipar_with_ndvi_maize_subset.csv', index=False)
```

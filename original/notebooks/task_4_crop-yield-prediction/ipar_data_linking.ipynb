This script is used to link the two sheets and add location information to the former:
- The dataset sheet that Margaux produced by cleaning up the IPAR data. Link: https://docs.google.com/spreadsheets/d/1nvWVdWE9L2dXTytEopdFmONm23y1vAKD8bF5F6EYMrc/edit?usp=sharing
- IPAR menage sheet which contains: "Department","Counties", "Village", "gpslat","gpslon". Link: https://docs.google.com/spreadsheets/d/1yUnvjFIdjEgtxUtkTaN-hq-duMEr4PD3Y7sec7guECQ/edit?usp=sharing 

The sheets are accessed by downloading them to the Github folder "../data/external/IPAR/"

New sheets are created in the same folder (file+'_updated') 

```python
import pandas as pd
import numpy as np
```

```python
# Good to do: work directly with the google sheets

ipar_file_baseline = pd.read_excel('../data/external/IPAR/yields_IPAR.xlsx', sheet_name='4_NM_SN_baseline_Production',encoding='utf-8-sig')
ipar_file_rice_CSC = pd.read_excel('../data/external/IPAR/yields_IPAR.xlsx', sheet_name='rice_CSC',encoding='utf-8-sig')
ipar_file_millet = pd.read_excel('../data/external/IPAR/yields_IPAR.xlsx', sheet_name='Millet',encoding='utf-8-sig')
ipar_file_corn = pd.read_excel('../data/external/IPAR/yields_IPAR.xlsx', sheet_name='Corn',encoding='utf-8-sig')
ipar_file_rice_irrigated_winter = pd.read_excel('../data/external/IPAR/yields_IPAR.xlsx', sheet_name='Rice irrigated winter',encoding='utf-8-sig')

#print(ipar_file_baseline)
print("Shape and head of ipar_file_baseline:")
print(ipar_file_baseline.shape)
print(ipar_file_baseline.head(3))

print("Shape and head of ipar_file_rice_CSC:")
print(ipar_file_rice_CSC.shape)
print(ipar_file_rice_CSC.head(3))


print("Shape and head of ipar_file_millet:")
print(ipar_file_millet.shape)
print(ipar_file_millet.head(3))

      
print("Shape and head of ipar_file_corn:")
print(ipar_file_corn.shape)
print(ipar_file_corn.head(3))

      
print("Shape and head of ipar_file_rice_irrigated_winter")
print(ipar_file_rice_irrigated_winter.shape)
print(ipar_file_rice_irrigated_winter.head(3))
      
ipar_menage_file = pd.read_excel('../data/external/IPAR/1_NM_SN__baseline_Menages.xlsx')
print("Shape and head of ipar_menage_file:")
print(ipar_menage_file.shape)
print(ipar_menage_file.head(3))
```

```python
# Confirm that the KEY is unique for the menage file
if len(ipar_menage_file["KEY"].unique()) == ipar_menage_file.shape[0]:
    print("Unique KEY for ipar_menage_file")
else:
    print("Key is not unique for ipar_menage_file")
```

```python
new_columns_list = ["KEY", "Department","Counties", "Village", "gpslat","gpslon"]
ipar_menage_file_subset_dataframe = ipar_menage_file.loc[:,new_columns_list] 
print(ipar_menage_file_subset_dataframe.head(3))
```

```python
ipar_file_rice_CSC_updated = ipar_file_rice_CSC.merge(ipar_menage_file_subset_dataframe, on='KEY',how='left')
print(ipar_file_rice_CSC_updated)
```

```python
ipar_file_baseline_updated = ipar_file_baseline.merge(ipar_menage_file_subset_dataframe, on='KEY',how='left')
print(ipar_file_baseline_updated)
```

```python
ipar_file_millet_updated = ipar_file_millet.merge(ipar_menage_file_subset_dataframe, on='KEY',how='left')
print(ipar_file_millet_updated)
```

```python
ipar_file_corn_updated = ipar_file_corn.merge(ipar_menage_file_subset_dataframe, on='KEY',how='left')
print(ipar_file_corn_updated)
```

```python
ipar_file_rice_irrigated_winter_updated = ipar_file_rice_irrigated_winter.merge(ipar_menage_file_subset_dataframe, on='KEY',how='left')
print(ipar_file_rice_irrigated_winter_updated)
```

```python
# Now write back to excel
# Write each updated dataframe to the sheets
writer = pd.ExcelWriter('../data/external/IPAR/yields_IPAR_updated.xlsx', engine='xlsxwriter')
ipar_file_baseline_updated.to_excel(writer, sheet_name='4_NM_SN_baseline_Production', index=False)
ipar_file_rice_CSC_updated.to_excel(writer, sheet_name='rice_CSC', index=False)
ipar_file_millet_updated.to_excel(writer, sheet_name='Millet', index=False)
ipar_file_corn_updated.to_excel(writer, sheet_name='Corn', index=False)
ipar_file_rice_irrigated_winter_updated.to_excel(writer, sheet_name='Rice irrigated winter', index=False)
writer.save()
```

```python
import os
import rasterio
```

```python
image_file = 'img_2002.tif' #test2 in code.earthengine.com

satdat = rasterio.open(image_file)
```

```python
print(satdat.count)
```

```python
# The dataset reports a band count.
print(satdat.count)

# And provides a sequence of band indexes.  These are one indexing, not zero indexing like Numpy arrays.
print(satdat.indexes)
```

```python
ndvi, evi = satdat.read(1), satdat.read(2)
```

```python
ndvi = ndvi*0.0001
```

```python
from matplotlib import pyplot as plt

fig = plt.imshow(ndvi)
fig.set_cmap('gist_earth')
plt.colorbar()

# Display the results.
plt.show()
```

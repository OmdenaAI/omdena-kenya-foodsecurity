```python
def create_new_df(df_func, weeks, weeks_to_month, year):
    row_list=[]
    df_func = df_func[df_func["year"]==year]
    num_locations = int(len(df_func)/weeks)
    print(num_locations)  
    for loc in range(num_locations):
        loc_df = df_func[loc*weeks:(loc+1)*weeks]
        #print(len(loc_df))
        #lat, long = loc_df["gpslat"][0], loc_df["gpslon"][0]
        
        for i in range(12):#12): #i-i+1
            avg_df = loc_df[weeks_to_month[str(i)]:weeks_to_month[str(i+1)]].mean()

            row_list.append({"year":year, "month":(i+1),"gpslat":avg_df["gpslat"], "gpslon":avg_df["gpslon"], "temp":avg_df["temp"],
                          "rel_hum":avg_df["rel_hum"], "hum":avg_df["hum"], "wind_speed":avg_df["wind_speed"], 
                           "wind_dir":avg_df["wind_dir"], "solar_radiat":avg_df["solar_radiat"],
                           "thermal_radiat":avg_df["thermal_radiat"], "dir_solar_radiat":avg_df["dir_solar_radiat"],
                          "soil_1":avg_df["soil_1"], "soil_2":avg_df["soil_2"], "soil_3":avg_df["soil_3"],
                           "soil_4":avg_df["soil_4"], "pr":avg_df["pr"]})
            
    return row_list
            
```

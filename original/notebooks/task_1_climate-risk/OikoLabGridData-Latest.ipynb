```python
import requests
import os
import json
import pandas as pd
```

```python
#A file containing your api key obtained from the website - https://oikolab.com/
with open("keyfile_tanya.txt") as f:
    api_key = f.readlines()[0].rstrip()
```

```python
r = requests.get('https://api.oikolab.com/weather',
                 params={'param': ['temperature',  
                                   'relative_humidity', 
                                    'humidex', 
                                    'surface_solar_radiation', 
                                    'surface_thermal_radiation', 
                                    'surface_direct_solar_radiation', 
                                    'wind_speed', 'wind_direction',
                                    'soil_temperature_level_1', 
                                    'soil_temperature_level_2', 
                                    'soil_temperature_level_3', 
                                    'soil_temperature_level_4', 
                                    'total_precipitation'],
                         'start': '2014-01-01',
                         'end': '2016-12-31',
                         'lat': 14.6937,
                         'lon': -17.44406,
                         'api-key': api_key}
                )
```

```python
weather_data = json.loads(r.json()['data'])

df = pd.DataFrame(index=pd.to_datetime(weather_data['index'], 
                                       unit='s'),
                  data=weather_data['data'],
                  columns=weather_data['columns'])
```

```python
df
```

```python
raw = df.rename(columns = {
      'temperature (degC)':'temp',
      'relative_humidity (0-1)':'rel_hum',
      'humidex (degC)':'hum',
      'wind_speed (m/s)':'wind_speed',
      'wind_direction (deg)':'wind_dir',
      'surface_solar_radiation (W/m^2)':'solar_radiat',
      'surface_thermal_radiation (W/m^2)':'thermal_radiat',
      'surface_direct_solar_radiation (W/m^2)':'dir_solar_radiat',
      'urban_temperature (degC)':'urban_temp',
      'soil_temperature_level_1 (degC)':'soil_1',
      'soil_temperature_level_2 (degC)':'soil_2',
      'soil_temperature_level_3 (degC)':'soil_3',
      'soil_temperature_level_4 (degC)':'soil_4',
      'total_precipitation (mm of water equivalent)':'pr'
  }) 
```

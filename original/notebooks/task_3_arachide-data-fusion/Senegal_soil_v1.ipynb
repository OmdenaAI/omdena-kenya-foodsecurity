```python
image_file = 'C:/Users/Faiz/Desktop/_Omdena/datasets/soilgrids/soilgrids_pH_central_5_15_mean.tif' 
phdat = rasterio.open(image_file)
```

```python
print(phdat.count)
```

```python
# The dataset reports a band count.
print(phdat.count)

# And provides a sequence of band indexes.  These are one indexing, not zero indexing like Numpy arrays.
print(phdat.indexes)
```

```python
ph_mean = phdat.read(1)
```

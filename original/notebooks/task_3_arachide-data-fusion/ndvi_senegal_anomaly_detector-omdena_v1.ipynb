```python
import matplotlib as mpl
import numpy as np
```

```python
def generatepng_annomaly(values):
    cmap = mpl.colors.ListedColormap(["#A50026","#000000", "#006837"])
    bounds = [-1,-0.0001,1]
    norm = mpl.colors.BoundaryNorm(bounds, cmap.N)
    plt.imshow(values, cmap=cmap, norm=norm)
    plt.axis('off')
    plt.show()
    return
```

```python
change1 = ndvi_2010 - ndvi_2002
generatepng_annomaly(change1)
```

```python
change2 = ndvi_2014 - ndvi_2010
generatepng_annomaly(change2)
```

```python
total = ndvi_2014 - ndvi_2002
generatepng_annomaly(total)
```
